```python
def isPrime(num):
  for i in range(2, (num//2) + 1):
    if(num % i == 0):
      return False

  return True
```

```python
def sumOfPrimes(num):
  sumArr = []
  sum = 0
  for j in range(2, num):
    if(isPrime(j)):
      sumArr.append(j)
      sum += j

  print(f"Prime numbers: {sumArr} less than or equal to {num}")
  print(f"Sum of prime numbers: {sum}")
```

```python
sumOfPrimes(10)
```

```python
def largeProductPairs(list):
  largestProduct = 0

  for i in range(0, len(list)-1):
    product = list[i] * list[i+1]
    if(product > largestProduct):
      largestProduct = product

  return largestProduct
```

```python
list  = [5, 6, 1, 0, 11, 0]

largeProductPairs(list)
```

```python

def validPassword(password):
  numeric = False
  upperCase = False
  lowerCase = False
  specialSymbol = False
  stringLenCheck = False

  specialChars = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '_', '-', '+', '=', '{', '}', '[', ']', '|', '\\', ':', ';', '"', "'", '<', '>', ',', '.', '?', '/']

  if(6<= len(password) <= 20):
    stringLenCheck = True

  for char in password:
    if(char.isnumeric()):
      numeric = True
    if(char.isupper()):
      upperCase = True
    if(char.islower()):
      lowerCase = True
    if(char in specialChars):
      specialSymbol = True


  if(stringLenCheck and numeric and upperCase and lowerCase and specialSymbol):
    print("Valid password!")

  if(not numeric):
    print("Invalid password! Your password needs to have at least one number!")

  if(not upperCase):
    print("Invalid password! Your password needs to have at least one uppercase letter!")

  if(not lowerCase):
    print("Invalid password! Your password needs to have at least one lowercase letter!")

  if(not specialSymbol):
    print("Invalid password! Your password needs to have at least one special symbol!")

  if(not stringLenCheck):
    print("Invalid password! Your password needs to be between 6 to 20 characters long!")

password = input("Input password")
validPassword("helasd")
```

```python
def strengthPassword(password):
  stringLenCheck = False
  newLineSpace = False
  consecutiveChar = False
  sameStringPattern = False

  list = []


  if(9 <= len(password) <= 20):
    stringLenCheck = True

  if "\n" or " " in password:
    newLineSpace = True

  for i in range(0, len(password)-3):
    if(password[i] ==  password[i+1] == password[i+2]):
      consecutiveChar = True

  for i in range(0, len(password)):
    for j in range(0, len(password)):
      list.append(password[i:j])
      if((j-i >=2) and (password[i:j] in list)):
        sameStringPattern = True

  if(stringLenCheck and not newLineSpace and not consecutiveChar and not sameStringPattern):
    print("Strong Password")

  if(not stringLenCheck):
     print("Invalid password! Your password needs to be between 9 to 20 characters long!")

  if(newLineSpace):
     print("Invalid password! Your password cannot have a newline or space character!")

  if(consecutiveChar):
     print("Invalid password! Your password cannot have more than two consecutive characters!")

  if(sameStringPattern):
     print("Invalid password! Your password cannot have the same string pattern (minimum 2 characters length)!")
```
